```python
import pandas as pd
import numpy as np

whey_protein = pd.read_csv(r'Supplement dataset.csv', sep=',', decimal=';', encoding='utf-8')

display(whey_protein)


# Função para formatar em dólar

def formatar(valor):
    valor_formatado = f"US${valor:,.2f}"
    return valor_formatado
```

```python
# Quais são todos os produtos de cada marca, e quantos produtos cada marca possui?

produtos_das_marcas_df = whey_protein.pivot(columns='Brand', values='Product Name')
produtos_marcas_df_nv = produtos_das_marcas_df.reset_index(drop=True)
display(produtos_marcas_df_nv.head())
```

```python

df_velha = whey_protein

# função pivot_table para transformar a tabela
df_nova = df_velha.pivot_table(index=df_velha.groupby('Brand').cumcount(), columns='Brand', values='Product Name', aggfunc=lambda x: ' '.join(str(v) for v in x))

# Resetando o índice para conseguir o formato desejado
df_nova.reset_index(drop=True, inplace=True)


for marca in df_nova.columns:
    c = 0
    for produto in df_nova[marca]:
        if pd.isna(produto):
            c +=0
        else:
            c += 1
    if c > 1:        
        print(f'A {marca} tem {c} produtos')
    else:
        print(f'A {marca} tem {c} produto')      
    
    
display(df_nova)
df_nova.info()
display(whey_protein)
```

```python
# Qual o melhor whey protein?

# 1 -- Mais proteina por dose de 30 gramas | min = 80%
whey_protein['Protein Percentage'] = (whey_protein['Protein (grams)'] / whey_protein['Serving Size (grams)']) * 100


melhor_whey = whey_protein[['Product Name', 'Serving Size (grams)', 'Protein (grams)', 'Protein Percentage', 'Calories', 'Carbs (grams)', 'Sugar (grams)', 'Sodium (milligrams)']].copy()

melhor_whey = melhor_whey.loc[melhor_whey['Protein Percentage'] >= 80]

# 2 -- Menos calorias | max = 1

melhor_whey = melhor_whey.loc[melhor_whey['Calories'] <= 120]
melhor_whey = melhor_whey.loc[melhor_whey['Sodium (milligrams)'] <= 120]
melhor_whey = melhor_whey.sort_values('Protein Percentage', ascending=False)

# formatei
def formatar(valor):
    valor_formatado = f"{valor:,.2f}%"
    return valor_formatado


melhor_whey['Protein Percentage'] = melhor_whey['Protein Percentage'].apply(formatar)


print('Os melhores Wheys Protein, são: ')
display(melhor_whey)
```
